# Demandas de ML

## ¿Cuáles son los valores promedio de alquiler por tipo de propiedad?

```python
# Ver el valor de los alquileres y calcular el promedio. Este es el valor total de las propiedades pero no por tipo de propiedad.
datos['Valor'].mean()
```

```python
# Agrupo los tipos de datos para poder sacar el promedio por tipo.
datos.groupby('Tipo').mean(numeric_only=True)
```

```python
# Calculo el valor promedio por tipo en formato Series
datos.groupby('Tipo')['Valor'].mean()
```

```python
# Calculo el valor promedio por tipo en formato DF y ordenarlos en forma ascendente por valor
datos.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
```

```python
# Almaceno los datos en una variable y grafico
# Grafico tipo barras horizontales, tamanio figura y el color
df_tipo_precio = datos.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
df_tipo_precio.plot(kind='barh', figsize=(12, 8), color='purple')
```

```python
# Mostramos cada tipo de inmueble con unique
datos.Tipo.unique()
```

```python
# Lista de inmuebles comerciales
inmuebles_comerciales = ['Conjunto Comercial/Sala', 'Edificio Completo', 'Tienda/Salón', 'Casa de Vecindad','Terreno Estándar', 'Cochera/Estacionamiento',
                         'Galpón/Depósito/Almacén', 'Tienda en Centro Comercial', 'Hotel', 'Loteo/Condominio', 'Industria', 'Casa Comercial']
```

```python
# Consultamos si alguno de los elementos de la lista de inmuebles comerciales esta dentro de nuestro data set
datos.query('@inmuebles_comerciales in Tipo')
```

```python
# Para ver los inmuebles residenciales
datos.query('@inmuebles_comerciales not in Tipo')
```

```python
# Almaceno los inmuebles residenciales en una variable ya que es lo que me piden de ML
df = datos.query('@inmuebles_comerciales not in Tipo')
df.head()
```

```python
# Calculamos los valores promedio por tipo de inmuebles residenciales
df_tipo_precio = df.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
df_tipo_precio.plot(kind='barh', figsize=(12, 8), color='purple')
```

## Porcentaje de cada tipo de inmueble presente en la base de datos

```python
# Unique de inmuebles residenciales
df.Tipo.unique()
```

```python
# Contamos la cantidad de cada uno de los inmuebles
df.Tipo.value_counts()
```

```python
# Porcentajes
df.Tipo.value_counts(normalize=True)
```

```python
porcentaje_tipo = df.Tipo.value_counts(normalize=True).to_frame()
porcentaje_tipo.plot(kind='bar', figsize=(12,8), color='green', xlabel='Tipos', ylabel='Porcentajes')
```
